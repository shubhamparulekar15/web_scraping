```
textSizes = {}
allPages = ""
for page in range(0, len(doc)):
    Page = doc.getPageText(page,output="json")
    blocks = json.loads(Page)["blocks"]
    for block in blocks:
        if "lines" in block:
            for line in block["lines"]:
                for span in line["spans"]:
                    
                    size = span["size"]
                    if size not in textSizes:
                        textSizes[size] = len(span["text"])
                    else:
                        textSizes[size] += len(span["text"])
```

```
validFontSize = heapq.nlargest(10, textSizes, key=textSizes.get)[0]
```

```
print (validFontSize)
```

```
actual_font_size = validFontSize
length = len (textSizes)
```

```
for i in range (0, length):
    numba = heapq.nlargest(10, textSizes, key=textSizes.get)[i]
    intnumba = int(numba)
    if (numba - intnumba) == 0 or (numba - intnumba) == 0.5:
        actual_font_size = numba
        break
```
