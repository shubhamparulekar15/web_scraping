```
import pandas as pd

df = pd.DataFrame(columns=["title","date","overview","download"])

pages= 2
for page in range(pages):
    sources = requests.get('https://www.weforum.org/reports/?page='+str(page+1)).text
    soup = BeautifulSoup(sources,'lxml')
    article = soup.find_all('article')
    print("")
    # print(article)
    for art in article:
        # print(art.a.h3.prettify())
        # print(art.a.h3.text)
        title = art.a.h3.text
        # print(title)
        date = art.find('div',class_='caption')
        date = date.text.replace("Report — ","")
        # overview = art.find('p').text
        # print(overview)
        reports_1 = art.find_all("a",{"class":"tout__link"})
        # print(reports_1[0].get("href"))

        art_link = "https://www.weforum.org"+reports_1[0].get("href")
        # print(art_link)

        sources_art = requests.get(art_link).text
        soup_art = BeautifulSoup(sources_art,'lxml')
        # st__content-block st__content-block--text
        overview = soup_art.find("div",{"class":"st__content-block st__content-block--text"})
        overview = overview.p.text

        pdf_link = soup_art.find("a",{"class":"report__link--pdf"})
        # print("---------------------------------------------------------------------")
        # print(title)
        # print(pdf_link)
        if(pdf_link):
            pl = pdf_link.get("href")
        else:
            pl = None
        # pdf_link = pdf_link.get("href")

        df = df.append({"title":title,"date":date,"overview":overview,"download":pl},ignore_index=True)



df.to_csv("output_web_scrape.csv",index = False)
```
